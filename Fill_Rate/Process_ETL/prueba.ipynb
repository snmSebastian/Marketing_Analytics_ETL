```python

# --- CONFIGURACION DE RUTAS ---
# Ruta de entrada para los archivos hist√≥ricos de ventas
path_data_historic = r'C:\Users\SSN0609\Stanley Black & Decker\Latin America - Regional Marketing - Marketing Analytics\Data\Raw\Fill Rate\prueba'
# Ruta de salida para los archivos Parquet procesados
path_data_processed = r'C:\Users\SSN0609\Stanley Black & Decker\Latin America - Regional Marketing - Marketing Analytics\Data\Processed-Dataflow\Fill Rate'


def group_parquet(input_path, output_path):
    """
    Lee archivos Excel de un directorio, los consolida y los guarda como
    archivos Parquet particionados por a√±o-mes.
    """

    # --- LECTURA Y CONSOLIDACION DE ARCHIVOS ---
    # Buscar todos los archivos .xlsx en el directorio de entrada.
    all_files_xlsx = glob.glob(os.path.join(input_path, "*.xlsx"))

    if not all_files_xlsx:
        print(f"Advertencia: No se encontraron archivos .xlsx en '{input_path}'.")
        return

    # Leer cada archivo y agregarlo a una lista de DataFrames:
    lst_files_xlsx = []
    for filename in all_files_xlsx:
        try:
            print(f"Leyendo archivo: {os.path.basename(filename)}")
            # Usar pd.read_excel para archivos .xlsx, no pd.read_csv
            df = pd.read_excel(filename, dtype=str, engine='openpyxl')
            # Eliminar la columna 'co' si existe, ignorando el error si no se encuentra.
            df = df.drop(columns=['co'], errors='ignore')
            lst_files_xlsx.append(df)
        except PermissionError:
            print(f"  [ERROR] Permiso denegado para leer el archivo: {os.path.basename(filename)}."
                  "\n  Aseg√∫rate de que no est√© abierto en Excel y vuelve a intentarlo.")
        except Exception as e:
            print(f"  [ERROR] No se pudo procesar el archivo {os.path.basename(filename)}: {e}")

    # Concatenar todos los DataFrames en uno solo
    df_consolidated = pd.concat(lst_files_xlsx, axis=0, ignore_index=True)

    

    # --- PROCESAMIENTO Y AGRUPACION ---
    # Crear una columna 'year_month' para usar en la agrupaci√≥n (ej: '2023-01').
    # Se usa .str.zfill(2) para asegurar que los meses tengan dos d√≠gitos (ej: '01', '02', etc.)
    # lo que mejora la consistencia y el orden de los nombres de archivo.
    try:
        df_consolidated['fk_year_month'] = (df_consolidated['Fiscal Year'].astype(str) + '-' +
                                       df_consolidated['Fiscal Period'].astype(str).str.zfill(2))
        df_consolidated['fk_date_country_customer_clasification'] = (df_consolidated['fk_year_month'] + '-' +
                                                    df_consolidated['Country Code']+ '-' +
                                                    df_consolidated['Sold-To-Customer']+ '-' +
                                                    df_consolidated['GPP Division']+ '-' +
                                                    df_consolidated['GPP Category']+ '-' +
                                                    df_consolidated['GPP Portfolio']).str.lower().str.strip()
        for col in df_consolidated.columns:
            df_consolidated[col] = df_consolidated[col].astype(str).str.upper()
    except KeyError as e:
        print(f"Error: La columna {e} no se encontr√≥ en los archivos. "
              "Aseg√∫rate de que las columnas 'Fiscal Year' y 'Fiscal Period' existan.")
        return

    # --- ESCRITURA DE ARCHIVOS PARQUET SEGMENTADOS ---
    # Agrupar el DataFrame por 'year_month' y guardar cada grupo en un archivo Parquet.
    for period, group in df_consolidated.groupby('fk_year_month'):
        # Crear un nombre de archivo descriptivo, ej: sales_2023-01.parquet
        output_filename = f"fill_rate{period}.parquet"
        output_full_path = os.path.join(output_path, output_filename)

        print(f"Guardando grupo {period} en: {output_full_path}")
        # Guardar el grupo en formato Parquet, excluyendo el √≠ndice.
        group.to_parquet(output_full_path, index=False)

    print("\nProceso completado. Archivos Parquet generados exitosamente.")

```

```python
# Librerias


def read_parquet(path):
    """
    Lee archivos Parquet de un directorio, los consolida y los guarda como
    archivos Parquet particionados por a√±o-mes.
    """
    
    # Buscar todos los archivos .parquet en el directorio de entrada.
    all_files_parquet = glob.glob(os.path.join(path, "*.parquet"))

    if not all_files_parquet:
        print(f"Advertencia: No se encontraron archivos .parquet en '{path}'.")
        return

    # Leer cada archivo y agregarlo a una lista de DataFrames:
    lst_files_parquet = []
    for filename in all_files_parquet:
        try:
            print(f"Leyendo archivo: {os.path.basename(filename)}")
            df = pd.read_parquet(filename)
            lst_files_parquet.append(df)
        except PermissionError:
            print(f"  [ERROR] Permiso denegado para leer el archivo: {os.path.basename(filename)}."
                  "\n  Aseg√∫rate de que no est√© abierto en otro programa y vuelve a intentarlo.")
        except Exception as e:
            print(f"  [ERROR] No se pudo procesar el archivo {os.path.basename(filename)}: {e}")

    # Concatenar todos los DataFrames en uno solo
    #df_consolidated = pd.concat(lst_files_parquet, axis=0, ignore_index=True)

    return lst_files_parquet
```

```python

def upload_files_update(path_files_update):
    """
    Lee archivos Excel de un directorio, los consolida y los guarda como
    archivos Parquet particionados por a√±o-mes.
    """

    # Buscar todos los archivos .xlsx en el directorio de entrada.
    all_files_xlsx = glob.glob(os.path.join(path_files_update, "*.xlsx"))

    if not all_files_xlsx:
        print(f"Advertencia: No se encontraron archivos .xlsx en '{path_files_update}'.")
        return

    # Leer cada archivo y agregarlo a una lista de DataFrames:
    lst_files_xlsx = []
    for filename in all_files_xlsx:
        try:
            print(f"Leyendo archivo: {os.path.basename(filename)}")
            df = pd.read_excel(filename,header=0)
            lst_files_xlsx.append(df)
        except PermissionError:
            print(f"  [ERROR] Permiso denegado para leer el archivo: {os.path.basename(filename)}."
                  "\n  Aseg√∫rate de que no est√© abierto en Excel y vuelve a intentarlo.")
        except Exception as e:
            print(f"  [ERROR] No se pudo procesar el archivo {os.path.basename(filename)}: {e}")



    # Concatenar todos los DataFrames en uno solo
    if len(lst_files_xlsx)>1:
        df_consolidated = pd.concat(lst_files_xlsx, axis=0, ignore_index=True)
    else:
        df_consolidated = lst_files_xlsx[0]

    df_consolidated['fk_year_month'] = (df_consolidated['Fiscal Year'].astype(str) + '-' +
                                       df_consolidated['Fiscal Period'].astype(str).str.zfill(2))
        
    df_consolidated['fk_date_country_customer_clasification'] = (df_consolidated['fk_year_month'] + '-' +
                                                    df_consolidated['Country Code']+ '-' +
                                                    df_consolidated['Sold-To-Customer']+ '-' +
                                                    df_consolidated['GPP Division']+ '-' +
                                                    df_consolidated['GPP Category']+ '-' +
                                                    df_consolidated['GPP Portfolio']).str.lower().str.strip()
    for col in df_consolidated.columns:
            df_consolidated[col] = df_consolidated[col].astype(str).str.upper()

    return df_consolidated
```
