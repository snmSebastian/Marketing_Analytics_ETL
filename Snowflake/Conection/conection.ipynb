# Conexion snowflake usando llave publica

```python
import snowflake.connector
from cryptography.hazmat.primitives import serialization
import os

def conectar_snowflake_keypair(ruta_llave, password_llave=None):
    """
    Establece una conexión con Snowflake utilizando autenticación de par de claves. Permitiendo asi saltar la validacion de clave externa
    
    :param ruta_llave: Ruta física al archivo rsa_key.p8
    :param password_llave: Password de la llave (str) o None si no tiene
    :return: Objeto de conexión (ctx) o None si falla
    """
    
    # Configuración de parámetros internos
    config = {
        "account": "PAA12529-SBD_CASPIAN",
        "user": "SEBASTIAN.NUNEZ@SBDINC.COM",
        "role": "PUBLIC",
        "warehouse": "DEV_AIDA_WH",
        "database": "PROD_MARTS",
        "schema": "DEMAND"
    }

    try:
        # 1. Carga y procesamiento de la llave privada
        with open(ruta_llave, "rb") as key_file:
            p_key = serialization.load_pem_private_key(
                key_file.read(),
                password=password_llave.encode() if password_llave else None
            )

        pkb = p_key.private_bytes(
            encoding=serialization.Encoding.DER,
            format=serialization.PrivateFormat.PKCS8,
            encryption_algorithm=serialization.NoEncryption(),
        )

        # 2. Establecer conexión
        ctx = snowflake.connector.connect(
            **config,
            private_key=pkb
        )
        
        print("Conexión exitosa mediante Key-Pair.")
        return ctx

    except Exception as e:
        print(f"Error al intentar conectar con Snowflake: {e}")
        return None

# --- EJEMPLO DE USO ---
if __name__ == "__main__":
    path = r"C:\Users\SSN0609\key snowflake\rsa_key.p8"
    
    conexion = conectar_snowflake_keypair(path)
    
    if conexion:
        try:
            cursor = conexion.cursor()
            cursor.execute("SELECT current_version()")
            print(f"Versión de Snowflake: {cursor.fetchone()[0]}")
        finally:
            conexion.close()
            print("Conexión cerrada.")
```
